```python
#import required libraries
import pandas as pd
```

```python
#load dataset from the drive and assign it to data variable
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Power BI/archive/telecom_customer_churn.csv')
```

```python
#to see first five rows of the dataset
data.head()
```

```python
#load second dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Power BI/archive/telecom_zipcode_population.csv')
```

```python
df.head()
```

```python
#merge those two csv files
merged_df = pd.merge(data, df, on='Zip Code',how='left')
```

```python
merged_df
```

```python
#checking whether if there is a null values or not
merged_df.isna().sum()
```

```python
# Fill missing values in 'Avg Monthly Long Distance Charges' column with the median value of the column
median_Avg_Monthly_Long_Distance_Charges = merged_df['Avg Monthly Long Distance Charges'].median()
merged_df['Avg Monthly Long Distance Charges'].fillna(median_Avg_Monthly_Long_Distance_Charges, inplace=True)
```

```python
merged_df.isna().sum()
```

```python
#Filling null values with previous value
merged_df1 =  merged_df.fillna(method = 'pad')
```

```python
merged_df1.isna().sum()
```

```python
#do feature engineering to fill missing values
merged_df2 = merged_df1.fillna({'Churn Category' : 'Other', 'Churn Reason':'Poor expertise of phone support'})
```

```python
merged_df2.isna().sum()
```

```python
#checks duplicates
merged_df2.duplicated().sum()
```

```python
#to save csv file
merged_df2.to_csv('preprocessed_merged_dataset.csv', index=False)
```
